```python
import json
import pandas as pd
import os
```

```python
os.getcwd()
```

```python
tweets_data = []
tweets_file_path = "./Classification-of-Depression-on-Social-Media-Using-Text-Mining/tweetdata.txt"
tweets_file = open(tweets_file_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except Exception as e:
        continue
len(tweets_data)
tweets_df = pd.DataFrame(tweets_data)
tweets_df.head(10)
```

```python
labels_df = pd.read_excel("./Classification-of-Depression-on-Social-Media-Using-Text-Mining/output.xlsx")
labels_df.head(10)
```

```python
labeled_df = tweets_df.merge(labels_df, on="id", suffixes=("_L", "_R"))
labeled_df.head(10)
```

```python
output_df = labeled_df[["id", "created_at", "text", "source", "sentiment"]]
# Rename column 'sentiment' to 'label' for better clarity
output_df.rename(columns={'sentiment':'label'}, inplace=True)
output_df.head(10)
```

```python
output_df.to_csv("./set1_processed.csv")
```
